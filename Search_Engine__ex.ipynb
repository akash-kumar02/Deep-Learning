```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# sample doc
doc = [
    "machine learning is part of AI.",
    "deep learning is a subset of machine learning.",
    "python is used for data science and machine learning.",
    "AI is and robotics are relateed fields."
]
```

```python
# step 1: convert text into tfidf vector
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(doc)
```

```python
# user query
query = "machine learning"
query_vector=vectorizer.transform([query])
```

```python
# step 3. compute cosine sim
similarity_score=cosine_similarity(query_vector,tfidf_matrix)
```

```python
import numpy as np
# step 4. rank doc
ranked_doc_indices = np.argsort(similarity_score[0])[::-1]
```

```python
# step 5. display res
print("search result:\n")
for index in ranked_doc_indices:
  print(f"score:{similarity_score[0][index]} - document: {doc[index]}")
  print(doc[index])
  print()
```

# Basic keyword search engine

```python
# sample doc
docs = [
     "machine learning is part of AI.",
     "deep learning is uses for neural network.",
     "python is used for data science and machine learning.",
     "AI is and machine learning are related fields."
]
```

```python
# search function
def search(query):
  results=[]
  for doc in docs:
    if query.lower() in doc.lower():
      results.append(doc)
  return results
```

```python
# ex
query = "machine learning"
results = search(query)
```

```python
print("search result:")
for r in results:
  print("-",r)
```
