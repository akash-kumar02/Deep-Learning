```python
documents = [
    "Data science is an interdisciplinary field",
    "Machine learning is a part of data science",
    "Deep learning is a subset of machine learning",
    "Python is widely used in data science",
    "Statistics is important for data analysis",
    "Neural networks are used in deep learning",
    "Big data requires efficient data processing",
    "Artificial intelligence uses machine learning",
    "Data visualization helps in data analysis",
    "Machine learning algorithms improve automatically"
]
```

# **Vectorization**

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
count_vect = CountVectorizer(stop_words='english')
X_count = count_vect.fit_transform(documents)
```

# **LDA (Latent Dirichlet Allocation)**

Probabilistic topic model (documents = mixtures of topics)

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X_count)
```

```python
words = count_vect.get_feature_names_out()

for i, topic in enumerate(lda.components_):
    print(f"Topic {i}:",
          [words[j] for j in topic.argsort()[-5:]])
```

# **LSA (Latent Semantic Analysis)**

M**atrix decomposition (SVD) → captures hidden semantic structure**

```python
from sklearn.decomposition import TruncatedSVD
```

```python
tfidf_vect = TfidfVectorizer(stop_words='english')
X_tfidf = tfidf_vect.fit_transform(documents)
```

```python
lsa = TruncatedSVD(n_components=2, random_state=42)
lsa.fit(X_tfidf)
```

```python
words = tfidf_vect.get_feature_names_out()

for i, topic in enumerate(lsa.components_):
    print(f"Topic {i}:",
          [words[j] for j in topic.argsort()[-5:]])
```

# **NMF (Non-Negative Matrix Factorization)**

**Parts-based topic model, often very interpretable**

```python
from sklearn.decomposition import NMF
```

```python
nmf = NMF(n_components=2, random_state=42)
nmf.fit(X_tfidf)
```

```python

for i, topic in enumerate(nmf.components_):
    print(f"Topic {i}:",
          [words[j] for j in topic.argsort()[-5:]])
```

```python
nmf = NMF(n_components=2, random_state=42)
W = nmf.fit_transform(X_tfidf)
H = nmf.components_
```

```python
feature_names = tfidf_vect.get_feature_names_out()
for topic_idx, topic in enumerate(H):
    print(f"\nTopic {topic_idx+1}:")
    top_words = [feature_names[i] for i in topic.argsort()[-5:]]
    print(top_words)
```
