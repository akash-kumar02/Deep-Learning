```python
# simple knowledge base
knowledge_base = {
    "what is machine learning?": "Machine learning is a subset of AI.",
    "what is deep learning?": "Deep learning is a subset of machine learning.",
    "what is reinforcement learning?": "Reinforcement learning is a subset of machine learning",
    "what is python?": "Python is a high-level programming language.",
}
def answer_question(question):
    question = question.lower()
    return knowledge_base.get(question, "I'm sorry, I don't know the answer to that question.")
```

```python
# example
print(answer_question("What is machine learning?"))
```

```python
print(answer_question("what is AI?"))
```

```python
# TF-IDF based q/a sys
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# sample doc
doc = ["Machine learning is a subset of AI.",
       "Deep learning is a subset of machine learning.",
       "Reinforcement learning is a subset of machine learning.",
       "Python is a high-level programming language.",
       "AI is the simulation of human intelligence."
]
```

```python
vect = TfidfVectorizer()
tfidf_matrix = vect.fit_transform(doc)

def answer_question(question):
    question_vector = vect.transform([question])
    similarity = cosine_similarity(question_vector, tfidf_matrix)
    index = similarity.argmax()
    return doc[index]
```

```python
# example
print(answer_question("What is machine learning?"))
```
