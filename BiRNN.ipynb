```python
# import libbraries

import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN, Bidirectional, Input
from tensorflow.keras.utils import pad_sequences
```

```python
# load dataset

vocab_size = 10000

(x_train, y_train),(x_test, y_test) = imdb.load_data(num_words=vocab_size)
```

```python
# padding sequence

max_len=100

x_train = pad_sequences(x_train, maxlen=max_len)
x_test = pad_sequences(x_test, maxlen=max_len)
```

```python
# SIMPLERNN MODEl

rnn_model = Sequential()

# Input Layer
rnn_model.add(Input(shape=(100,)))

# Embedding Layer
rnn_model.add(Embedding(input_dim=vocab_size,output_dim= 128))

# RNN Layer
rnn_model.add(SimpleRNN(64))

# Output Layer
rnn_model.add(Dense(1, activation='sigmoid'))

#compile model
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# show model summary
rnn_model.summary()
```

```python
# train model
# y_train = y_train[1]
# y_test = y_test[1]

history_rnn= rnn_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_split=0.2)


# Evaluate model
loss_rnn, accuracy_rnn = rnn_model.evaluate(x_test, y_test)
print(f'Test Loss: {loss_rnn}, Test Accuracy: {accuracy_rnn}')
```

```python
# bidirectional Rnn Layer
birnn_model = Sequential()
birnn_model.add(Input(shape=(max_len,))) # Use max_len for input shape
birnn_model.add(Embedding(input_dim=vocab_size, output_dim=128))
birnn_model.add(Bidirectional(SimpleRNN(64)))

# Output Layer
birnn_model.add(Dense(1, activation='sigmoid'))

#compile model
birnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# show model summary
birnn_model.summary()
```

```python
# train model

history_birnn= birnn_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_split=0.2)

# evaluate model

loss_birnn, accuracy_birnn = birnn_model.evaluate(x_test, y_test)
print(f'Test Loss: {loss_birnn}, Test Accuracy: {accuracy_birnn}')
```
