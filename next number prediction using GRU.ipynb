```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
```

```python
sequence = np.array([i for i in range (1,201)])
```

```python
#2. prepare input-output pairs

X, y = [] , []

time_steps = 5

for i in range(len(sequence)- time_steps):
  X.append(sequence[i:i + time_steps])
  y.append(sequence[i+time_steps])

X = np.array(X)
y = np.array(y)
```

```python
#3. reshape input for GRU (sample , time steps , feature)

X = X.reshape((X.shape[0] , X.shape[1] , 1))
```

```python
#4. build GRU model


model = Sequential()
model.add(GRU(50 , activation = 'relu' , input_shape = (X.shape[1] , X.shape[2])))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam' , loss='mse')
```

```python
model.fit(X,y  ,epochs= 100 , verbose =0)
```

```python
test_input = np.array([ 98, 99, 100,101,102])

test_input = test_input.reshape((1, time_steps , 1))



predication = model.predict(test_input , verbose = 0)
print(predication)
```
