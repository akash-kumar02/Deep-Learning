```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
doc=["Data science includes machine learning",
     "Machine learning algorithms build models",
     "Deep learning uses neural network",
     "Neural nettworks are used in deep learning",
     "Football match was exciting",
     "TRhe team won the football game"]
```

```python

vectorizer=TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(doc)
```

```python
# Apply NMF
nmf = NMF(n_components=2, random_state=42)
W = nmf.fit_transform(X)
H = nmf.components_
```

n_components=2 =nomber of topics

w==documents=topic matrix

h=topic=word matrix

```python
# display topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(H):
    print(f"\nTopic {topic_idx+1}:")
    top_words = [feature_names[i] for i in topic.argsort()[-6:]]
    print(top_words)
```
