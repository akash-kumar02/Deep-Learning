```python
# import libraries

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU
```

```python
#1 Generate Sequence

sequence = np.array([i for i in range(1,201)])
```

```python
#2 prepare input-output pairs

x,y = [],[]
time_steps= 5

for i in range (len(sequence)-time_steps):
  x.append(sequence[i:i+time_steps])
  y.append(sequence[i+time_steps])

x = np.array(x)
y = np.array(y)
```

```python
y
```

```python
#3 reshape input for GRU (samples, time steos, features)

x= x.reshape((x.shape[0],x.shape[1],1))
```

```python
x
```

```python
#4 buil GRU model
model= Sequential()
model.add(GRU(50,activation='relu',input_shape=(time_steps,1)))
model.add(Dense(1))
```

```python
#5 compile model

model.compile(optimizer='adam',loss='mse')
```

```python
#6 train model

model.fit(x,y,epochs= 100, verbose=0)
```

```python
#7 test presdictions

test_input= np.array([196,197,198,199,200])
test_input= test_input.reshape(1,time_steps,1)
model.predict(test_input,verbose=0)
```
