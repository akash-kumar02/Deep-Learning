## 🤔 Why Pre-trained Embeddings?

Training Word2Vec from scratch needs **a lot of text**.  
Instead, we can download embeddings already trained on **billions of words**!

| Model | Made by | Trained on | Special power |
|-------|---------|-----------|---------------|
| **GloVe** | Stanford | 6 billion words (Wikipedia) | Great analogies & similarity |
| **FastText** | Facebook AI | Wikipedia + Web crawl | Handles **unknown words** (OOV)! |

## ❓ What is OOV?
**Out-Of-Vocabulary (OOV)** = a word the model has never seen before.

- GloVe → ❌ Cannot handle OOV words  
- FastText → ✅ Builds a vector from **character sub-words** (n-grams)

## 🔢 Part A — GloVe

### A1: Converting Words to Vectors

Every word maps to a fixed-length array of numbers.  
Words with **similar meanings** → similar arrays.


````python
print("GloVe — Word Vectors (first 8 of 50 numbers)\n")

for word in ['king', 'queen', 'man', 'woman', 'cat']:
    vec = glove[word]
    preview = ', '.join([f'{v:.3f}' for v in vec[:8]])
    print(f"  '{word:8s}' →  [{preview} ...]")
````

### A2: Out-of-Vocabulary (OOV) Words in GloVe

GloVe has a fixed dictionary. If a word isn't in it → **no vector**.  
We write a safe helper function to handle this gracefully.


````python
def get_glove_vector(model, word):
    """
    Safely retrieve a GloVe vector.
    Returns the vector if found, or None if the word is OOV.
    """
    w = word.lower()      # GloVe uses lowercase
    if w in model:
        return model[w]   # ✅ found
    return None           # ❌ OOV — GloVe can't help us here


# Test with known and unknown words
test_words = ['king', 'smartphone', 'ChatGPT', 'dog', 'xyzabc123']

print("Testing GloVe OOV handling:\n")
for w in test_words:
    v = get_glove_vector(glove, w)
    if v is not None:
        print(f"  ✅ '{w}'  →  found!   (first value: {v[0]:.4f})")
    else:
        print(f"  ❌ '{w}'  →  OOV — GloVe has no vector!")

print("""
💡 OOV Strategies for GloVe:
   1. Return None / a zero vector
   2. Average the vectors of other known words in the same sentence
   3. Switch to FastText — it handles OOV natively! (see Part B)
""")
````

### A3: Word Similarity with GloVe

GloVe computes similarity using **cosine similarity** between vectors.

```
Similarity = cos(θ) = (A · B) / (|A| × |B|)
```

| Score | Meaning |
|-------|---------|
| `1.0` | Identical meaning |
| `0.5` | Related |
| `0.0` | Unrelated |
| `-1.0` | Opposite meaning |


````python
pairs = [
    ('king',  'queen'),
    ('man',   'woman'),
    ('cat',   'dog'),
    ('paris', 'france'),
    ('good',  'bad'),
    ('king',  'banana'),
]

print(f"  {'Word Pair':<22} {'Similarity':>10}  Relationship")
print(f"  {'-'*22}  {'-'*10}  {'-'*20}")

for w1, w2 in pairs:
    try:
        s = glove.similarity(w1, w2)
        if s > 0.7:   rel = "🟢 Very similar"
        elif s > 0.3: rel = "🟡 Somewhat related"
        else:          rel = "🔴 Not similar"
        print(f"  {w1+' & '+w2:<22}  {s:>10.4f}  {rel}")
    except KeyError as e:
        print(f"  {w1+' & '+w2:<22}  ⚠️  not found: {e}")
````

### A4: Analogy Tasks with GloVe ✨

The famous **king − man + woman = queen** trick!

**Formula:**
```
A : B  ::  C : ?
? ≈ vector(B) − vector(A) + vector(C)
```

**Intuition:**
```
king   − man   =  "royalty vector"  (removes maleness)
queen  − woman =  "royalty vector"  (same!)

So:  king − man + woman  ≈  queen  🎉
```


````python
def solve_analogy(model, a, b, c, tag=""):
    """
    Solve:  a : b  ::  c : ?
    Math:   most similar to  vector(b) - vector(a) + vector(c)
    """
    label = f"[{tag}] " if tag else ""
    print(f"  {label}'{a}' : '{b}'  ::  '{c}' : ___?")
    try:
        results = model.most_similar(positive=[b, c], negative=[a], topn=3)
        for rank, (word, score) in enumerate(results, 1):
            print(f"    {rank}. '{word}'  (score: {score:.4f})")
    except KeyError as e:
        print(f"    ⚠️  Word not found: {e}")
    print()


analogies = [
    ('man',   'king',   'woman'),    # → queen
    ('paris', 'france', 'berlin'),   # → germany
    ('good',  'better', 'bad'),      # → worse
    ('king',  'prince', 'queen'),    # → princess
]

print("🎯 GloVe Analogy Results\n")
for a, b, c in analogies:
    solve_analogy(glove, a, b, c, "GloVe")
````

## 🚀 Part B — FastText Embeddings

### What makes FastText special?

FastText breaks every word into **character n-grams** (sub-word pieces):

```
'playing'  →  '<pl', 'pla', 'lay', 'ayi', 'yin', 'ing', 'ng>', ...
```

Each piece has its own vector → sum them up → vector for **any** word!

```
'unhappiness' (OOV) → 'un' + 'hap' + 'app' + 'ppi' + ... → ✅ works!
```


````python
def build_demo_fasttext():
    """Small Word2Vec model that demonstrates FastText concepts offline."""
    sentences = [
        "the cat sat on the mat".split(),
        "dogs are loyal animals".split(),
        "the king rules the kingdom".split(),
        "the queen rules beside the king".split(),
        "a man walked into the forest".split(),
        "the woman followed the man".split(),
        "the lion is the king of the jungle".split(),
        "animals live in different habitats".split(),
        "dolphins are intelligent animals".split(),
        "tigers are endangered animals".split(),
        "good better bad worse sentiment".split(),
        "paris is the capital of france".split(),
        "berlin is the capital of germany".split(),
        "man king woman queen prince princess".split(),
    ]
    m = Word2Vec(sentences, vector_size=50, window=3,
                 min_count=1, sg=1, epochs=300, seed=42)
    return m.wv


def load_fasttext():
    """Try real FastText download; fall back to demo."""
    try:
        import gensim.downloader as api
        print("   Downloading FastText (~958 MB) — this takes a few minutes …")
        m = api.load("fasttext-wiki-news-subwords-300")
        print(f"   ✅ Real FastText loaded!  Vocab: {len(m):,}  Dim: {m.vector_size}")
        return m, False
    except Exception as e:
        print(f"   ⚠️  Download unavailable. Using demo model.")
        return build_demo_fasttext(), True


print("⏬ Loading FastText …")
ft, ft_demo = load_fasttext()
if ft_demo:
    print("\n   [DEMO MODE] Small Word2Vec + character n-gram OOV approximation.")
    print("   Download link for real FastText:")
    print("   https://dl.fbaipublicfiles.com/fasttext/vectors-english/\n")

print(f"   📐 Vector size : {ft.vector_size}")
print(f"   📚 Vocabulary  : {len(ft):,} words")
````

### B1: Converting Words to Vectors (FastText)

````python
print("FastText — Word Vectors (first 8 numbers)\n")

for word in ['king', 'queen', 'man', 'woman', 'cat']:
    if word in ft:
        vec = ft[word]
        preview = ', '.join([f'{v:.3f}' for v in vec[:8]])
        print(f"  '{word:8s}' →  [{preview} ...]")
    else:
        print(f"  '{word}' → not directly in demo vocabulary")
````

### B2: FastText OOV Magic — Character N-grams

Let's implement FastText's sub-word trick ourselves!


````python
def char_ngrams(word, min_n=3, max_n=6):
    """
    Break a word into character n-grams (sub-word pieces).
    FastText adds < > boundary markers.

    Example: 'cat' → '<ca', 'cat', 'at>'
    """
    padded = f"<{word}>"
    ngrams = []
    for n in range(min_n, max_n + 1):
        for i in range(len(padded) - n + 1):
            ngrams.append(padded[i:i+n])
    return ngrams


def get_fasttext_vector(model, word):
    """
    Get a vector for ANY word — even OOV!

    Steps:
      1. Check if word is directly in vocabulary  →  exact match
      2. Break into character n-grams, average known ones  →  approximation
      3. Return zero vector if nothing works  →  fallback
    """
    w = word.lower()

    # Step 1: exact lookup
    if w in model:
        return model[w], "exact match ✅"

    # Step 2: character n-gram approximation (FastText's trick!)
    ngrams = char_ngrams(w)
    known  = [ng for ng in ngrams if ng in model]
    if known:
        vectors = np.array([model[ng] for ng in known])
        return vectors.mean(axis=0).astype(np.float32), f"n-gram avg ({len(known)}/{len(ngrams)} pieces known) ✅"

    # Step 3: fallback
    return np.zeros(model.vector_size, dtype=np.float32), "zero vector ❌"


# ── Show the n-grams for a word ─────────────────────────────────────
demo_word = "playing"
ngrams = char_ngrams(demo_word)
print(f"Character n-grams of '{demo_word}':")
print(f"  {ngrams}\n")
````

````python
# ── Compare GloVe vs FastText OOV handling ─────────────────────────
oov_words = ['king', 'animals', 'unhappiness', 'playfulness',
             'ChatGPT', 'xyzabc123', 'running', 'walked']

print(f"  {'Word':<20}  {'GloVe':>16}  FastText")
print(f"  {'-'*20}  {'-'*16}  {'-'*35}")

for w in oov_words:
    g = "✅ Known" if get_glove_vector(glove, w) is not None else "❌ OOV"
    _, f_method = get_fasttext_vector(ft, w)
    print(f"  {w:<20}  {g:>16}  {f_method}")

print("""
🔑 Key takeaway:
   GloVe:    'unhappiness' → ❌ OOV (has no vector at all!)
   FastText: 'unhappiness' → ✅ built from sub-word pieces
""")
````

### B3: Word Similarity — GloVe vs FastText

````python
def cosine_similarity(v1, v2):
    """Manually compute cosine similarity between two vectors."""
    n1, n2 = np.linalg.norm(v1), np.linalg.norm(v2)
    if n1 == 0 or n2 == 0:
        return 0.0
    return float(np.dot(v1, v2) / (n1 * n2))


print(f"  {'Word Pair':<22} {'GloVe':>8}  {'FastText':>8}  Agreement")
print(f"  {'-'*22}  {'-'*8}  {'-'*8}  {'-'*12}")

for w1, w2 in pairs:
    try:
        g_sc = glove.similarity(w1, w2)
    except KeyError:
        g_sc = None

    v1, _ = get_fasttext_vector(ft, w1)
    v2, _ = get_fasttext_vector(ft, w2)
    f_sc = cosine_similarity(v1, v2)

    if g_sc is not None:
        agree = "✅ Agree" if abs(g_sc - f_sc) < 0.25 else "⚠️  Differ"
        print(f"  {w1+' & '+w2:<22}  {g_sc:>8.4f}  {f_sc:>8.4f}  {agree}")
    else:
        print(f"  {w1+' & '+w2:<22}  {'N/A':>8}  {f_sc:>8.4f}")
````
