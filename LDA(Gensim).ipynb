```python
documents = [
    "Data science combines statistics and programming.",
    "Machine learning improves prediction accuracy.",
    "Deep learning uses neural networks.",
    "Regular exercise keeps the body healthy.",
    "A balanced diet improves immunity.",
    "Football is played worldwide.",
    "Students prepare hard for examinations."]
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
stop_words = set(stopwords.words('english'))

texts =[]
for doc in documents:
    tokens = word_tokenize(doc.lower())
    tokens =[word for word in tokens if word.isalpha() and word not in stop_words]
    texts.append(tokens)
```

```python
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
lda_model = gensim.models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=2,
                                   passes=10)
```

```python
for topic in lda_model.print_topics():
    print(topic)
```
